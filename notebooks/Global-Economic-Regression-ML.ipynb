```python
# these datas have many outlier variable and they are normally distributed,so  decided delete outlier variables to reach better distribution.
#That's why our r2 scores are low. 
# Dataset is very noisy.
#To get higher r2 score  should cut outlier side of data and decrease in noisy. 
#I asked to ChatGPT to remove outliers and it printed these code(below cell).
```

```python
df_clean = df_copy.copy()

for col in df_clean.select_dtypes(include=np.number).columns:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
```

```python
print(df_copy.shape)
print(df_clean.shape)
```

```python
# After run this code we loss more than half of our data. 
# Macrovariable datas include many extereme variables such as high inflation rates n Turkiye or Argantina.
#These datas are correct and happend n real world,terms of economy we should not skip these extreme variables to make correct economic interpretation.
#To improve prediction of regression  should have delete these extereme variables.
# In addition now we can not visualize our data beacuse of we lost half of them. Just I wanted to warn you if you try to run code.
```

```python
X=df_clean[["Unemployment Rate (%)","Interest Rate (Real, %)","Inflation (GDP Deflator, %)","Current Account Balance (% GDP)"]]
y=df_clean["GDP Growth (% Annual)"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=19)
```

```python
Scaler=RobustScaler()
X_train_scaled=Scaler.fit_transform(X_train)
X_test_scaled=Scaler.transform(X_test)
```

```python
Grad=GradientBoostingRegressor()
Grad.fit(X_train_scaled,y_train)
y_pred=Grad.predict(X_test_scaled)
print(r2_score(y_test,y_pred))
```

```python
#now our prediction are more correct :))) 
#r2 score increased from 0.04 to 0.40 :)))
```

# Model Selection

```python
# In this part we select best model for our data 
# I call other regressors 
```

```python
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor,RandomForestRegressor
from sklearn.linear_model import Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
def evalute_model(true,predict):
    mse=mean_squared_error(true,predict)
    mae=mean_absolute_error(true,predict)
    R2_score=r2_score(true,predict)
    return mse,mae,R2_score
```

```python
models={
    "Linear Regression":LinearRegression(),
    "Lasso":Lasso(),
    "Ridge":Ridge(),
    "AdaBoost Regressor":AdaBoostRegressor(),
    "Gradient Boost Regressor":GradientBoostingRegressor(),
    "Random Forest Regressor":RandomForestRegressor(),
    "Decision Tree Regressor":DecisionTreeRegressor(),
    "KNeighbors Regressor":KNeighborsRegressor(),
    "XGBoost Regressor":XGBRegressor(),
    }
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model=model.fit(X_train_scaled,y_train)
    
    y_pred_train=model.predict(X_train_scaled)
    y_pred_test=model.predict(X_test_scaled)
    
    train_mse,train_mae,train_r2_score=evalute_model(y_train,y_pred_train)
    test_mse,test_mae,test_r2_score=evalute_model(y_test,y_pred_test)
    
    print(list(models.keys())[i])
    print("Model performance for Training Set")
    print("Mean Squared Error",train_mse)
    print("Mean Absolute Error",train_mae)
    print("r2 Score",train_r2_score)
    
    print("-------------------------")
    
    print("Model performance for Test Set")
    print("Mean Squared Error",test_mse)
    print("Mean Absolute Error",test_mae)
    print("r2 Score",test_r2_score)
    
    print("-------------------------")
    print("\n")
```

```python
# I am going to select Random Forest Regressor which has 0.45 r2 score (the highest r2_score)
# Regressors may learn well train data set for example  Decision Tree Regressor but when prediction time comes they could mistake 
# r2 score for training set of Decision Tree regressor is 0.96 (overfitting?!?!?!) overlearning 
#r2 score for test set of Decision Tree Regressor is 0.08, bad prediction
```

# Hyperparameter Tuning Part

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#this function using for hyperparameter tuning
#I read documentation of Random Forest Regressor to learn necessary parameters on google (its website)
#website:
#https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html
```

```python
rf_params={
    "n_estimators": [200,500,800],
    "min_samples_split": [2,3,5],
    "min_samples_leaf": [1,2],
}
```

```python
random_cv_model=[("Random Forest Regressor",RandomForestRegressor(),rf_params)]
```

```python
for name,model,params in random_cv_model:
    random=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,n_iter=100,n_jobs=-1,scoring="r2")
    random.fit(X_train_scaled,y_train)
    print("Best Params Of",name,random.best_params_)
```

```python
#We got best params for Random Forest Regressor 
#Now I refit this regressor using these parameters 
```
